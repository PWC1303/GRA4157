## Exercise 1: Simple Calculator

**Objective:** Build a simple calculator that performs basic arithmetic operations.

**Task:** Write a function called `simple_calculator()` that:

1. Asks the user for two numbers.
2. Asks the user to choose an operation (+, -, *, /).
3. Performs the operation and displays the result.

**Requirements:**

* Handle invalid inputs gracefully by displaying appropriate error messages.
* Do not proceed until valid inputs are provided.

### Bonus:

1. Implement more advanced operations like square root, power, and modulo.
2. Add an option to perform another calculation or exit.

```python
import operator
def simple_calculator():
    status = "again"
    while status == "again":
        ops = { "+": operator.add, "-": operator.sub, "*": operator.mul,"**": operator.pow, "/":operator.truediv, "modulo": operator.mod} 
        try:
            print("Please enter two numbers")
            n1 = int(input("Enter the first number and press enter"))
            n2 = int(input("Enter the second number and press enter"))
        except: ValueError
        print("Invalid inputs, enter only numbers ")
        try:
            operation = input("Please choose a an operation, type (+,-,*,/,root)")
            if operation == "root":
            
                result= n1**(1/n2)
            
                print("Invalid number combinatioon")
            else:
                result = ops[operation](n1,n2)
            print(f"{n1} {operation} {n2} = {result}")
        except KeyError:
                print("Invalid operation, try again input only (+,-,*,**,/,root,modulo,root)")
        status = input("Type exit to exit and again to try calculate again")
   

    
```

## Exercise 2: String Reverser

**Objective:** Understand string manipulations in Python.

**Task:** Write a function called `reverse_string()` that:

1. Asks the user for a string.
2. Reverses the string and prints it.

**Requirements:**

* The function should handle empty strings and whitespace.
* Provide an option to reverse another string or exit.

### Bonus:

1. Reverse words in the string while keeping their order (e.g., "hello world" becomes "olleh dlrow").

```python
def str_to_list_to_str_reverser(some_str):
    str_list = []
    rev_str_list=[]
    for i in some_str:
        str_list.append(i)

    for i in range(len(str_list)):
        rev_str_list.append(str_list[-i-1]) 

    rev_str = "".join(rev_str_list)

    return rev_str


def reverse_string():
    status = "again"
    while status == "again":
        some_str = input("give me a string baby")
        str_list = some_str.split()
        rev_str_list =[]
        for i in range(len(str_list)):
            str_i = str_list[i]
            rev_str_list.append(str_to_list_to_str_reverser(str_i))
        reversed_string = " ".join(rev_str_list)
        print(reversed_string)
        status = input("want to give another string, if so type again if not type exit ")

print(reverse_string())

    

```
