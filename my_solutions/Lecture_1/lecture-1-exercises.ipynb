### Task 3: File Extension Counts

1. Write another function named `count_file_extensions` that accepts a list of file names.
2. This function should return a dictionary where the keys are file extensions (e.g., ".txt", ".pdf") and the values are counts of files with those extensions.
3. File extensions should be considered case-insensitively (e.g., ".TXT" and ".txt" should be counted together).

````python
def count_file_extensions(file_name_ls):
    for element in file_name_ls:
        extensions = ["."+element.rsplit(".",1)[-1].lower() for element in file_name_ls]
    res = list(set(extensions))
    count_dict = dict.fromkeys(res,None)

    for unique in res:
        count = 0
        for element in extensions:
            if unique ==element:
                count = count +1
        
        count_dict[unique]=count
    return(count_dict)

listed_files = list_files(r"C:\Users\pvwol\Downloads")
count_file_extensions(listed_files)
   
````
